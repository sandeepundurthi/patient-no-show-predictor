```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('healthcare_noshows_appt.csv')
```

```python
print(df.shape)
df.head()
```

```python
# 4. Convert dates to datetime format
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
```

```python
df
```

```python
# 5. Add engineered features
df['AppointmentWeekday'] = df['AppointmentDay'].dt.day_name()
df['ScheduledWeekday'] = df['ScheduledDay'].dt.day_name()
df['LeadTime'] = (df['AppointmentDay'] - df['ScheduledDay']).dt.days
df['HourScheduled'] = df['ScheduledDay'].dt.hour
```

```python
# 6. Rename columns for clarity
df.rename(columns={
    'Hipertension': 'Hypertension',
    'Handcap': 'Handicap',
    'Showed_up': 'ShowedUp',
    'Date.diff': 'LeadTimeDays'
}, inplace=True)
```

```python
# 7. Clean incorrect ages
df = df[df['Age'] >= 0]  # remove negative ages
```

```python
# 8. Drop unnecessary columns
df.drop(['PatientId', 'AppointmentID'], axis=1, inplace=True)
```

```python
# 9. Reset index
df.reset_index(drop=True, inplace=True)
```

```python
# 10. Save cleaned data
df.to_csv('cleaned_appointments.csv', index=False)
```
